### **Теория**  
**1. Цикл `while`**  
**Цикл** — это конструкция, которая повторяет блок кода, пока выполняется условие.  
**Синтаксис:**  
```python
while условие:
    # код, который выполняется, пока условие истинно
```

**Пример:**  
```python
count = 5
while count > 0:
    print(f"Осталось: {count} сек")
    count -= 1  # Уменьшаем count на 1
print("Старт!")
```
**Вывод:**  
```
Осталось: 5 сек  
Осталось: 4 сек  
...  
Старт!
```

**Условие выхода:**  
Цикл завершится, когда `условие` станет ложным. Если условие всегда истинно, цикл станет **бесконечным** (это ошибка!).

---

**2. Цикл `for`**  
Цикл `for` используется для итерации (перебора) элементов коллекций: списков, строк, диапазонов и т.д.  
**Синтаксис:**  
```python
for элемент in коллекция:
    # код для каждого элемента
```

**Примеры:**  
- **Список:**  
  ```python
  fruits = ["яблоко", "банан", "вишня"]
  for fruit in fruits:
      print(fruit)
  ```
  Вывод: `яблоко`, `банан`, `вишня`.

- **Строка:**  
  ```python
  word = "Привет"
  for letter in word:
      print(letter)  # Выведет каждую букву отдельно.
  ```

- **Диапазон `range()`:**  
  ```python
  for i in range(3):  # 0, 1, 2
      print(i)
  ```

---

**3. Бесконечные циклы и их избегание**  
**Бесконечный цикл** — цикл, который никогда не завершается, потому что условие всегда истинно.  
**Пример ошибки:**  
```python
count = 5
while count > 0:
    print(count)  # Забыли уменьшить count → цикл не завершится!
```

**Как избежать:**  
- Всегда проверяйте, что условие цикла может стать ложным.  
- Используйте `break`, чтобы принудительно выйти из цикла.  
- Для `while` убедитесь, что переменная в условии меняется внутри цикла.

---

**4. Итерация по словарям и множествам**  
- **Словарь (dict):**  
  Перебираем ключи, значения или пары ключ-значение.  
  ```python
  student = {"имя": "Аня", "возраст": 20, "курс": 3}

  for key in student:          # Перебор ключей
      print(key)

  for value in student.values():  # Перебор значений
      print(value)

  for key, value in student.items():  # Перебор пар
      print(f"{key}: {value}")
  ```

- **Множество (set):**  
  Перебираем элементы в произвольном порядке.  
  ```python
  colors = {"красный", "синий", "зелёный"}
  for color in colors:
      print(color)
  ```

### **Практика**  
**1. Подсчёт суммы чисел от 1 до N**  
```python
n = int(input("Введите N: "))
total = 0
for num in range(1, n + 1):
    total += num
print(f"Сумма: {total}")
```

**2. Поиск факториала числа**  
```python
n = int(input("Введите число: "))
factorial = 1
for i in range(1, n + 1):
    factorial *= i
print(f"Факториал {n} = {factorial}")
```

**3. Вывод таблицы умножения**  
```python
number = int(input("Введите число: "))
for i in range(1, 11):
    print(f"{number} x {i} = {number * i}")
```

**4. Подсчёт гласных букв в строке**  
```python
text = input("Введите текст: ").lower()
vowels = "аеёиоуыэюя"
count = 0
for letter in text:
    if letter in vowels:
        count += 1
print(f"Гласных букв: {count}")
```

### **Домашнее задание**  
**1. Найти все простые числа до N**  
**Простое число** — число больше 1, которое делится только на 1 и само себя.  
**Пример решения:**  
```python
n = int(input("Введите N: "))
print("Простые числа:")
for num in range(2, n + 1):
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
```

**2. Поиск палиндромов в списке слов**  
**Палиндром** — слово, которое читается одинаково слева направо и справа налево (например, "топот").  
```python
words = ["топот", "машина", "казак", "радар", "привет"]
palindromes = []
for word in words:
    if word == word[::-1]:  # Сравниваем слово с его перевёрнутой версией
        palindromes.append(word)
print("Палиндромы:", palindromes)
```
